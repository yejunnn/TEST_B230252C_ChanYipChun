```python
import pandas as pd
import matplotlib.pyplot as plt


coffee_data = pd.read_csv("coffee.csv")


coffee_data.info()
```

```python

total_orders = coffee_data.groupby(['product_category', 'product_type'])['transaction_qty'].sum().reset_index()


total_orders.columns = ['category', 'product', 'total_order']


print("Total orders for each product:")
print(total_orders)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(total_orders['product'], total_orders['total_order'], color='skyblue')
plt.xlabel('Product')
plt.ylabel('Total Orders')
plt.title('Total Orders by Product Type')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
